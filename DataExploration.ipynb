```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc={"figure.figsize":(9,6)})
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
nrow = data.shape[0]
ncol = data.shape[1]
```

```python
def Pnull(data, col):
    return (data[col].isnull().sum()/nrow)

def data_info(data, col):
    print ('The first few records: ')
    print (data[col].head())
    print ('The percentage of null records: ')
    print (Pnull(data, col))
    
def boxplot(data, col):
    sns.boxplot(x=col, y='SalePrice', data=data)
```
