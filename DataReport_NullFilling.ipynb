```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df = pd.concat([df_train, df_test])
df.head()
```

```python
print('The dimension of the training dataset is: ', df_train.shape)
print('The dimension of the testing dataset is: ', df_test.shape)
```

```python
df['BsmtCond'].dtype == object
```

```python
df.columns
```

```python
quants = []
qualits = []

for col in df.columns:
    if df[col].dtype == object:
        qualits.append(col)
    else:
        quants.append(col)

print ('The quantatitive variables are: ')
print (quants)
print ()
print ('The qualitative variables are: ')
print (qualits)
```

```python
quants_nulls = df[quants].isnull().sum()
quants_nulls.loc[quants_nulls>0]
```

```python
quants_null_cols = quants_nulls.loc[quants_nulls>0].index
df[quants_null_cols].head()
```

```python
temp = list(quants_null_cols)
temp.remove('SalePrice')
```

```python
#LotFrontage - mean
temp = list(quants_null_cols)
temp.remove('SalePrice')
temp.remove('LotFrontage')
df.loc[:, temp] = df[temp].fillna(0)
```

```python
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
```

```python
df[quants_null_cols].isnull().sum()
```

```python
qualits_nulls = df[qualits].isnull().sum()
qualits_nulls.loc[qualits_nulls>0]
```

```python
qualits_null_cols = qualits_nulls.loc[qualits_nulls>0].index
df[qualits_null_cols].head()
```

```python
df[qualits_null_cols] = df[qualits_null_cols].fillna('None')
```

```python
df.isnull().sum()
```

```python
df.to_csv('NoNullData.csv')
```
